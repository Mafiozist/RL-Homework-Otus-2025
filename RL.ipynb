### –ú–∞—Ä–∫–æ–≤—Å–∫–∞—è —Ñ–æ—Ä–º–∞–ª–∏–∑–∞—Ü–∏—è (MDP) –∏ —Ü–∏–∫–ª –∞–≥–µ–Ω—Ç‚Äì—Å—Ä–µ–¥–∞
–ë–∞–∑–æ–≤–∞—è –º–æ–¥–µ–ª—å ‚Äî [MDP](https://incompleteideas.net/book/RLbook2020.pdf) $\langle S,A,P,R,\gamma\rangle$:

### –¢–µ–æ—Ä–∏—è (—Ä–∞—Å—à–∏—Ä–µ–Ω–æ)
- –ü–æ–ª–∏—Ç–∏–∫–∞: $\pi(a|s)$ ‚Äî —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –¥–µ–π—Å—Ç–≤–∏–π –≤ —Å–æ—Å—Ç–æ—è–Ω–∏–∏ $s$.
- –ü–µ—Ä–µ—Ö–æ–¥—ã –∏ –Ω–∞–≥—Ä–∞–¥—ã: $P(s'|s,a)$, $R(s,a,s')$; –¥–∏—Å–∫–æ–Ω—Ç $\gamma\in[0,1)$.
- Value –ø–æ–¥ –ø–æ–ª–∏—Ç–∏–∫–æ–π: 
  $$ v_\pi(s) = \sum_a \pi(a|s) \sum_{s'} P(s'|s,a)[R(s,a,s') + \gamma\, v_\pi(s')] $$
- Action‚Äëvalue: 
  $$ q_\pi(s,a) = \sum_{s'} P(s'|s,a)[R(s,a,s') + \gamma\, \sum_{a'}\pi(a'|s') q_\pi(s',a')] $$
- –°–≤—è–∑—å $v_\pi$ –∏ $q_\pi$: $v_\pi(s)=\sum_a\pi(a|s)q_\pi(s,a)$.

### –ü—Ä–∏–º–µ—Ä—ã –∏ —Ä–∞–∑–±–æ—Ä—ã
1) –¢–∞–±–ª–∏—á–Ω–∞—è —Å—Ä–µ–¥–∞ (–¥–µ—Ç–µ—Ä–º–∏–Ω–∏—Ä–æ–≤–∞–Ω–Ω–∞—è):
- –ò–∑ A –¥–µ–π—Å—Ç–≤–∏–µ–º Right ‚Üí B (r=0), –∏–∑ B —Ç–µ—Ä–º–∏–Ω–∞–ª (+1). –¢–æ–≥–¥–∞ $v(A)=\gamma\,v(B)=\gamma$.
- –ï—Å–ª–∏ –µ—Å—Ç—å –≤–µ—Ç–≤–ª–µ–Ω–∏–µ A‚Üí{B,C} —Å –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—è–º–∏ $p,1-p$, $v(A)=\gamma[p\,v(B)+(1-p)\,v(C)]$.

2) –°—Ç–æ—Ö–∞—Å—Ç–∏—á–µ—Å–∫–∞—è —Å—Ä–µ–¥–∞ (–º–∞—Ç. –æ–∂–∏–¥–∞–Ω–∏–µ):
- –ü—É—Å—Ç—å $P(B|A,Right)=0.8$, $P(C|A,Right)=0.2$, $v(B)=1, v(C)=0$, $\gamma=0.9$.
- –¢–æ–≥–¥–∞ $v(A)=0.9\,[0.8\cdot 1 + 0.2\cdot 0]=0.72$.
