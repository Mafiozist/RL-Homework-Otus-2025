# Описание задачи Taxi-v3

В этой среде есть четырe заранее определённых мест для посадки/высадки пассажира — красное, зелёное, жёлтое и синее — на поле размером 5 × 5 клеток.
Такси начинает в случайной клетке, пассажир находится в одном из этих четырёх мест.  

Цель агента:
1) Довезти такси до места, где пассажир находится;
2) Выполнить действие «подъём» (pickup);
3) Затем доставить пассажира к его целевому месту высадки;
4) Выполнить действие «высадка» (drop-off). После этого эпизод заканчивается.

# Пространство действий (Action Space)

0. Движение вниз (юг)
1. Движение вверх (север)
2. Движение вправо (восток)
3. Движение влево (запад)
4. Подъём пассажира (pickup)
5. Высадка пассажира (drop-off)

# Пространство наблюдений (Observation Space)

Всего дискретных состояний — 500:  
- 25 возможных позиций такси (5×5)
- 5 возможных положений пассажира (4 стартовых места + “в такси”)
- 4 возможных целевых мест высадки 

Наблюдение возвращается как целое число, кодирующее состояние.

# Награды (Rewards)

- За каждый шаг (если другое вознаграждение не получено) выдаётся −1
- За успешную доставку пассажира на правильное место: +20
- За незаконный подъём или высадку (в неправильной позиции): −10

# Конец эпизода (Episode end)

- Такси успешно высадило пассажира
- Либо исчерпан лимит шагов, если используется обёртка с ограничением по времени.
