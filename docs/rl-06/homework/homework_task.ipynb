# Задача LunaLander

Lunar Lander — это классическая задача оптимизации траектории ракеты. Согласно принципу максимума Понтрягина, оптимально либо полностью включать двигатель, либо выключать его. Поэтому в данной среде используются дискретные действия: двигатель включен или выключен.

## Пространство действий

Доступны четыре дискретных действия:
- **0**: ничего не делать
- **1**: включить левый двигатель ориентации
- **2**: включить основной двигатель
- **3**: включить правый двигатель ориентации

## Пространство наблюдений

Состояние представляет собой 8-мерный вектор:
- координаты лунного модуля по осям `x` и `y`
- линейные скорости по осям `x` и `y`
- угол наклона модуля
- угловая скорость
- два булевых значения, показывающие, касается ли каждая из посадочных опор поверхности

## Система наград

Награда начисляется на каждом шаге. Общая награда эпизода — это сумма наград за все шаги.

За каждый шаг награда:
- увеличивается/уменьшается, чем ближе/дальше модуль к посадочной площадке
- увеличивается/уменьшается, чем медленнее/быстрее движется модуль
- уменьшается, чем больше наклонен модуль (отклонение от горизонтали)
- увеличивается на 10 очков за каждую опору, касающуюся земли
- уменьшается на 0.03 очка за каждый кадр, когда работает боковой двигатель
- уменьшается на 0.3 очка за каждый кадр, когда работает основной двигатель

Эпизод получает дополнительную награду -100 или +100 очков за аварию или успешную посадку соответственно.

Эпизод считается решенным, если набрано не менее 200 очков.

## Начальное состояние

Модуль начинает в верхней центральной части экрана с случайной начальной силой, приложенной к его центру масс. Посадочная площадка всегда находится в координатах (0, 0).

## Завершение эпизода

Эпизод завершается, если:
1. модуль разбивается (тело модуля соприкасается с поверхностью Луны)
2. модуль выходит за пределы экрана (координата `x` больше 1)
3. модуль "засыпает" (перестает двигаться и не сталкивается с другими объектами)

Статус "Принято" ставится при выполнении всех критериев.
