# Задача Car Racing  

Car Racing — это задача обучения с подкреплением, в которой агент должен управлять гоночным автомобилем и проехать как можно дальше по случайно сгенерированной трассе, двигаясь по дороге и избегая выезда за её пределы.  

![Езда на машинах](./car_racing.gif)

## Пространство действий  

Действия являются непрерывными и задаются трёхмерным вектором:  
- управление рулём в диапазоне [-1, 1] (влево / вправо)  
- газ в диапазоне [0, 1]  
- тормоз в диапазоне [0, 1]  

## Пространство наблюдений  

Наблюдение представляет собой RGB-изображение размера 96×96 пикселей, полученное с камеры, закреплённой над автомобилем. Изображение содержит:  
- трассу  
- положение автомобиля относительно дороги  
- окружающую среду  

## Система наград  

Награда начисляется на каждом шаге. Общая награда эпизода — это сумма наград за все шаги.  

За каждый шаг:  
- агент получает положительную награду за продвижение по трассе  
- агент получает штраф за выезд за пределы дороги  
- агент получает небольшой штраф за каждый шаг, чтобы стимулировать быстрое прохождение трассы  

Если автомобиль длительное время не продвигается вперёд, эпизод считается неудачным.  

Эпизод считается решённым, если средняя награда за эпизод превышает 900 очков.  

## Начальное состояние  

Автомобиль появляется в начальной точке случайно сгенерированной кольцевой трассы. Конфигурация трассы (повороты, длина сегментов) генерируется заново для каждого эпизода.  

## Завершение эпизода  

Эпизод завершается, если:  
1. автомобиль надолго выезжает за пределы трассы  
2. превышено максимальное количество шагов эпизода  
3. автомобиль практически перестаёт двигаться
